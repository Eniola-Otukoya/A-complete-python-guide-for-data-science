## Filtering records with .query()

## Handling missing values using bfill and ffill

```python
sample = [[np.nan, 2, np.nan, 0],
         [3, 4, np.nan, 1],
         [np.nan, np.nan, np.nan, np.nan],
         [np.nan, 3, np.nan, 4]]
print(sample)
```

```python
type(sample)
```

```python
sample_df = pd.DataFrame(sample, columns = ['A', 'B', 'C', 'D'])
sample_df
```

```python
# to fill the NaN using bfill
sample_df['A'].fillna(method = "bfill", inplace = True)
sample_df
```

```python
# to fill the NaN using ffill
sample_df['A'].fillna(method = "ffill", inplace = True)
sample_df
```

```python
sample_df['C'].fillna(0, inplace = True)
sample_df
```

## Using Advanced Technique to handle missing values
- K-Nearest Neighbour (KNN) imputer: To use the KNN imputer from the scikit-learn library to impute missing values using the K-Nearest Neighbour approach.

```python
sample1 = [[np.nan, 2, 3, 0],
         [3, 4, np.nan, 1],
         [2, 4, np.nan, 5],
         [np.nan, 3, 3.5, 4],
         [4, 5, np.nan, 6]]
print(sample1)
```

```python
type(sample1)
```

```python
sample1_df = pd.DataFrame(sample1, columns = ["A", "B", "C", "D"])
sample1_df
```

```python
from sklearn.impute import KNNImputer

# Initialize the KNN algorithm
knn_imputer = KNNImputer()

# fit and transform the dataframe
sample1_df = knn_imputer.fit_transform(sample1)
sample1_df
```

```python
# to view in DataFrame
df = pd.DataFrame(sample1_df, columns = ["A", "B", "C", "D"])
df
```
